```python
import numpy as np
from scipy.integrate import odeint, solve_ivp
import matplotlib.pyplot as plt
```

```python
#PROBLEMA 1
def osc(y, t, z, w0):
    x, dxdt = y
    dydt = -2*z*w0*dxdt - (w0**2)*x
    return [dxdt, dydt]
#definimos el tiempo en el q integraremos
t = np.linspace(0,4*np.pi,1000)
# condiciones iniciales
t0 = 0.0 
y0 = [-1.0, 0.0]
w0 = np.sqrt(9/1)
#integramos en diferentes valores para z
sol0 = odeint(osc, y0, t, args=(0,w0,))
sol1 = odeint(osc, y0, t, args=(0.1,w0,))
sol2 = odeint(osc, y0, t, args=(0.2,w0,))
sol3 = odeint(osc, y0, t, args=(0.4,w0,))
sol4 = odeint(osc, y0, t, args=(0.7,w0,))
sol5 = odeint(osc, y0, t, args=(1,w0,))
sol6 = odeint(osc, y0, t, args=(2,w0,))
#graficamos
plt.figure(figsize=(20,10))
plt.grid()
plt.title("Problema 1")
plt.xlabel('wt')
plt.ylabel('y(t)')
plt.plot(t, sol0[:,0], label='z = 0')
plt.plot(t, sol1[:,0], label='z = 0.1')
plt.plot(t, sol2[:,0], label='z = 0.2')
plt.plot(t, sol3[:,0], label='z = 0.4')
plt.plot(t, sol4[:,0], label='z = 0.7')
plt.plot(t, sol5[:,0], label='z = 1.0')
plt.plot(t, sol6[:,0], label='z = 2.0')
plt.legend(loc='best')
```

```python
#PROBLEMA 2
def osc(y, t, z, w0, omega):
    x, dxdt = y
    dydt = -2*z*w0*dxdt - (w0**2)*x + np.sin(omega*t)
    return [dxdt, dydt]
#definimos el tiempo en el q integraremos
t = np.linspace(0,50,10000)
# condiciones iniciales
t0 = 0.0 
y0 = [0.0, 0.0]
z=0
w0=np.sqrt(9/1)
#integramos en diferentes valores para z
sol0 = odeint(osc, y0, t, args=(z,w0,3,))
sol1 = odeint(osc, y0, t, args=(z,w0,3.1,))
sol2 = odeint(osc, y0, t, args=(z,w0,3.3,))
sol3 = odeint(osc, y0, t, args=(z,w0,3.4,))
#graficamos
fig, axs = plt.subplots(4,figsize=(20,10))
fig.suptitle('Pulsos y resonancia')
axs[0].plot(t, sol0[:,0], label='omega = 3')
axs[1].plot(t, sol1[:,0], label='omega = 3.1')
axs[2].plot(t, sol2[:,0], label='omega = 3.3')
axs[3].plot(t, sol3[:,0], label='omega = 3.4')
plt.xlabel('t')
axs[0].legend(loc='best');
axs[1].legend(loc='best');
axs[2].legend(loc='best');
axs[3].legend(loc='best');
plt.show()
```
